Here, we'll be looking into the avg SAT scores across the states, answering the following question.

###**Q1: What are the average SAT scores for each states, and which state has the highest average score?**

```python
%%bigquery --project $project_id
SELECT
  state,region,
  ROUND(AVG(med_sat_value)) AS mean_sat,
  RANK() OVER (ORDER BY AVG(med_sat_value) DESC) AS ranking,
  ROUND(AVG(grad_100_value), 2) AS avg_grad100_rate,
  ROUND(AVG(grad_150_value), 2) AS avg_grad150_rate
FROM
  `ba-775-a06.cc_datasets.1-details`
WHERE
  med_sat_value IS NOT NULL AND grad_100_value IS NOT NULL AND grad_150_value IS NOT NULL
GROUP BY
  state,region
ORDER BY
  ranking;
```

**Q1 ANSWER**

D.C., Rhode Island, Utah, Wyoming, and Massachusetts have the highest average SAT scores, around 1100, and they are mostly located in the Northeast and West regions. Universities in the West region generally have lower graduation rates than those in the Northeast region.
North Caolina, Hawaii, Mississippi, West Virginia, Delaware have the lowest average SAT scores, around 980, and they are mostly located in the Southeast regions, and their graduation rates are generally lower than states with higher average SAT scores.

```python
%%bigquery --project $project_id
WITH CollegeRanks AS (
   SELECT
       state,
       chronname,
       med_sat_value,
       RANK() OVER (PARTITION BY state ORDER BY med_sat_value DESC) AS sat_rank
   FROM
       `ba-775-a06.cc_datasets.1-details`
   WHERE
       med_sat_value IS NOT NULL  -- Exclude NULL values if necessary
)


SELECT
   t.state,
   t.chronname AS flagship_college,
   t.med_sat_value AS flagship_sat_value,
   r.chronname AS college_with_highest_sat,
   r.med_sat_value AS highest_sat_value
FROM
   `ba-775-a06.cc_datasets.1-details` t
JOIN
   CollegeRanks r ON t.state = r.state AND r.sat_rank = 1
WHERE
   t.flagship = 'X'  -- Filter for flagship colleges
   AND t.med_sat_value IS NOT NULL
ORDER BY
   flagship_sat_value DESC,t.state, t.chronname;
```

###**Q3: How does the comparison between awards and flagship colleges unfold?**

```python
%%bigquery --project $project_id

WITH FlagshipComparison AS (
    SELECT
        state,
        AVG(CASE WHEN flagship = 'X' THEN awards_per_value END) AS flagship_awards
    FROM
        `ba-775-a06.cc_datasets.1-details`
    WHERE
        awards_per_value IS NOT NULL
    GROUP BY
        state
)

SELECT
    t.state,
    AVG(t.awards_per_value) AS avg_awards_all_colleges,
    fc.flagship_awards AS awards_flagship,
    fc.flagship_awards - AVG(t.awards_per_value) AS awards_difference
FROM
    `ba-775-a06.cc_datasets.1-details` t
JOIN
    FlagshipComparison fc ON t.state = fc.state
WHERE
    t.awards_per_value IS NOT NULL AND fc.flagship_awards IS NOT NULL
GROUP BY
    t.state, fc.flagship_awards
ORDER BY
    awards_flagship DESC, t.state;
```

**Q3 ANSWER**

The table reveals a discernible pattern that could offer valuable insights in favor of flagship universities over others. Surprisingly, the awards value for the flagship university exceeded the average awards value across the state.
